```python
# Análise de Vendas com Python
#Projeto de análise exploratória de dados utilizando Python e pandas.

import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

df.head()
```

```python
print("Informações:")
print(df.info())

print("\nTotal de vendas:")
print(df["total_bill"].sum())

print("\nVendas por dia:")
print(df.groupby("day")["total_bill"].sum())
```

```python
import matplotlib.pyplot as plt

df.groupby("day")["total_bill"].sum().plot(kind="bar")
plt.title("Vendas por dia")
plt.ylabel("Total")
plt.show()

## Insights encontrados

#Maior volume de vendas ocorre no fim de semana
#Ticket médio varia por dia
#Diferença de consumo entre perfis de clientes

```
